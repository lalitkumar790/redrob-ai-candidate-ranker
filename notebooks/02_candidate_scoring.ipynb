```python
scores = []

for candidate in candidates[:1000]:
    score = calculate_signal_score(
        candidate["redrob_signals"]
    )
    scores.append(score)

print("Min:", min(scores))
print("Max:", max(scores))
print("Average:", sum(scores)/len(scores))
```

```python
github_scores = []
profile_scores = []
search_scores = []
saved_scores = []

for candidate in candidates[:5000]:

    signals = candidate["redrob_signals"]

    github_scores.append(
        signals.get("github_activity_score", 0)
    )

    profile_scores.append(
        signals.get("profile_completeness_score", 0)
    )

    search_scores.append(
        signals.get("search_appearance_30d", 0)
    )

    saved_scores.append(
        signals.get("saved_by_recruiters_30d", 0)
    )

print("GitHub")
print(min(github_scores), max(github_scores))

print("Profile")
print(min(profile_scores), max(profile_scores))

print("Search")
print(min(search_scores), max(search_scores))

print("Saved")
print(min(saved_scores), max(saved_scores))
```
